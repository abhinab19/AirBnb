```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('compressed_data.csv')
```

```python
df.head()
```

```python
df.columns
```

# Checking for missing values

```python
print(df.isnull().sum())
```

```python
df.info()
```

# Handling Missing Values

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per  month': 0, 'last review': df['last review'].min()}, inplace = True)
```

```python
df.dropna(subset = ['NAME', 'host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns = ['license', 'house_rules'], errors = 'ignore')
```

```python
df.head()
```

```python
df['price'] = df['price'].replace('[\$,]', '', regex = True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex = True).astype(float)
```

```python
df.head()
```

# Removing Duplicates

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

```python
df.describe()
```

# Visulization

# What is the distribution of listing price ?

```python
plt.figure(figsize = (10,6))
sns.histplot(df['price'], bins = 50, kde = True, color = 'red')
plt.title("Distribution of listing price.")
plt.xlabel("Price $")
plt.ylabel("Frequency")
plt.show()
```

The histrogram shows a fairly even distribution of 
listing prices across different price ranges, indicating no
particular concerntration of listings in any specific price range.
The KDE line helps visualize this even spread more clearly,
confirming that the dataset contains listings with a wide variety of prices.

# How different room types are distributed?

```python
df['room type']
```

```python
plt.figure(figsize = (8,5))
sns.countplot(x= "room type", data = df, color = 'blue')
plt.title("room type distributed")
plt.xlabel("room type")
plt.ylabel("count")
plt.show()
```

# How are the listings distributed across different neighbourhoods?

```python
plt.figure(figsize = (12,8))
sns.countplot(y='neighbourhood group', data=df, color = 'lightgreen', order=df['neighbourhood group'].value_counts().index)
plt.title('Number of Listings distributed across the neighbourhood')
plt.xlabel('Count')
plt.ylabel('Neighbourhood group')
plt.show()
```

# What is the relationship between room type and price?

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="room type", y="price" , hue = "room type", data=df, palette = 'Set1' )
plt.title('Room type vs Price')
plt.xlabel('Room type')
plt.ylabel('Price $')
plt.legend(title='Room Type')
plt.show()
```

# How has the reviews changed over time?

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line', color='red')
plt.title('Number of reviews over time')
plt.xlabel('Date')
plt.xlabel('Number of reviews')
plt.show()
```
