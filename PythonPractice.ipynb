```python
class Mathops:
  def add(self, a=None, b=None, c=None):

    if a is not None and b is not None and c is not None:
      return a+b+c
    elif a is not None and b is not None:
      return a+b
    return a

obj = Mathops()
print(obj.add(2,3,5))
print(obj.add(2,3))
print(obj.add(5))
```

```python
class Parent:
  def __init__(self, name):
    self.name = name
    print("Parent Constructor called")

class Child(Parent):
  def __init__(self, name, age):
    super().__init__(name) # Fixed: added parentheses to super()
    self.age = age
    print("Child constructor called")

C = Child("Abhinab", 25)
print(C.name, C.age)
```

```python
#__getattribute__
class Demo:
  def __init__(self):
    self.x = 10
  def __getattribute__(self, name):
    print(f"Accessing {name}")
    return super().__getattribute__(name)

d = Demo()
print(d.x)
```

```python
#__setattr__
class Demo:
  def __setattr_(self, name, value):
    print(f"Setting {name} = {value}")
    super().__setattr__(name, value)

d = Demo()
d.x = 42
print(d.x)
```

```python
#Abstraction
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Square(Shape):
    def __init__(self, side):
        self.side = side
    def area(self):
        return self.side ** 2

sq = Square(5)
print(sq.area())  # 25
```

```python
#Strategy Dessign
def CreditCard(amount): print(f"Paid {amount} with credit card")
def PayPal(amount): print(f"Paid {amount} with Pay pal")

class PaymentContext:
  def __init__(self, strategy):
    self.strategy = strategy
  def set_strategy(self, strategy):
    self.strategy = strategy
  def pay( self, amount):
    self.strategy(amount)

context = PaymentContext(CreditCard)
context.pay(100)

context.set_strategy(PayPal)
context.pay(200)

```

```python
#Factory Design Pattern
class Dog():
  def speak(self):
    return "Woof!"

class Cat():
  def speak(self):
    return "Meow!"


class AnimalFactory():
  @staticmethod

  def create_animal(animal_type : str):
    if animal_type == "dog":
      return Dog()
    elif animal_type == "cat":
      return Cat()
    else:
      return ValueError("Unlnown animal type")


animal = AnimalFactory.create_animal("dog")
print(animal.speak())

animal = AnimalFactory.create_animal("cat")
print(animal.speak())
```

```python
#Singleton Design Pattern
class Singleton:
  _instance = None

  def __new__(cls, *args, **kwargs):
    if not cls._instance:
      cls._instance = super().__new__(cls)

    return cls._instance

s1 = Singleton()
s2 = Singleton()

print(s1 is s2)
```

```python
#Decorator Design Pattern
from abc import ABC, abstractmethod

class Coffee(ABC):
  @abstractmethod
  def cost(self): pass

class SimpleCoffee(Coffee):
  def cost(self):
    return 50

class MilkDecorator(Coffee):
   def __init__(self,coffee):
    self._coffee = coffee
   def cost(self):
     return self._coffee.cost() + 20 # Adding cost of milf

class SugarDecorator(Coffee):
  def __init__(self, coffee):
    self._coffee = coffee

  def cost(self):
    return self._coffee.cost() + 10 #Adding cost sugar

coffee = SimpleCoffee()
print(coffee.cost())

coffee = MilkDecorator(coffee)
print(coffee.cost())

coffee = SugarDecorator(coffee)
print(coffee.cost())
```

```python
#Pub/Sub Design Pattern
from collections import defaultdict

class PubSub:
  def __init__(self):
    self.subscribers = defaultdict(list)
  def subscribe(self, event_type, callback):
    self.subscribers[event_type].append(callback)
  def publish(self, event_type, data):
    for callback in self.subscribers[event_type]:
      callback(data)

def handle_order(data):
  print(f"Order received: {data}")

def send_email(data):
  print(f"Email sent for order: {data}")

bus = PubSub()
bus.subscribe("order_created", handle_order)
bus.subscribe("order_created", send_email)

bus.publish("order_created", {'id': 101, 'item': 'Laptop'})
```

```python
#Facade Design Pattern
# Subsystem classes
class CPU:
    def freeze(self):
        print("CPU freeze")
    def jump(self, position):
        print(f"CPU jump to {position}")
    def execute(self):
        print("CPU execute")

class Memory:
    def load(self, position, data):
        print(f"Loading '{data}' into memory at {position}")

class HardDrive:
    def read(self, sector, size):
        return f"Data from sector {sector}, size {size}"

# Facade
class ComputerFacade:
    def __init__(self):
        self.cpu = CPU()
        self.memory = Memory()
        self.hd = HardDrive()

    def start(self):
        self.cpu.freeze()
        data = self.hd.read(10, 1024)
        self.memory.load(0, data)
        self.cpu.jump(0)
        self.cpu.execute()

# Usage
computer = ComputerFacade()
computer.start()   # âœ… One simple call
```

```python
#Adapter Design Pattern
class EuropeanSocket:
  def voltage(self): return 230
  def frequency(self): return 50
  def plug_type(self): return "Type C"

class IndianPlug:
  def plug_in(self):
    return "Plugged in to Indian socket"

class SocketAdapter:
  def __init__(self, european_socket):
    self._european_socket = european_socket
  def plug_in(self):
    return (f"Adapted: {self._european_socket.voltage()}V, "
    f"{self._european_socket.frequency()}HZ, "
    f"Plug type: {self._european_socket.plug_type()}")

euro_socket = EuropeanSocket()
adpater = SocketAdapter(euro_socket)
print(adpater.plug_in())
```

```python
#Closure using a decorator
def decorator(func):
  def Wrapper():
    print("Before function call")

    func()
    print("After function call")
  return Wrapper

@decorator
def greet():
  print("hello")
greet()
```
